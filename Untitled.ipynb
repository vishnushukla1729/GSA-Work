# Comparison on Q9 between f (first-year) and t (transfer) 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure we have the latest data
q9f_data = pd.read_csv('q9f_data.csv', encoding='ascii')
q9t_data = pd.read_csv('q9t_data.csv', encoding='ascii')

def clean_transfer_data(df):
    df_cleaned = df.dropna(how='all')
    df_cleaned.reset_index(drop=True, inplace=True)
    return df_cleaned

df_cleaned = clean_transfer_data(q9t_data)

def calculate_satisfaction_percentages(df):
    total_responses = df.shape[0] * df.shape[1]
    satisfaction_counts = df.melt()['value'].value_counts()
    satisfaction_percentages = (satisfaction_counts / total_responses * 100).round(2)
    return satisfaction_percentages

q9f_percentages = calculate_satisfaction_percentages(q9f_data.iloc[:, 1:])
q9t_percentages = calculate_satisfaction_percentages(df_cleaned.iloc[:, 1:])

comparison_df = pd.DataFrame({'First-Year': q9f_percentages, 'Transfer': q9t_percentages}).fillna(0)

def calculate_total_satisfaction(df):
    satisfied_responses = df.isin(['Very Satisfied', 'Satisfied']).sum().sum()
    total_responses = df.size
    satisfaction_percentage = (satisfied_responses / total_responses) * 100
    return satisfaction_percentage
first_year_satisfaction = calculate_total_satisfaction(q9f_data.iloc[:, 1:])

transfer_satisfaction = calculate_total_satisfaction(df_cleaned.iloc[:, 1:])
satisfaction_df = pd.DataFrame({
    'Group': ['First-Year', 'Transfer'],
    'Satisfaction Percentage': [first_year_satisfaction, transfer_satisfaction]
})
def calculate_satisfaction_breakdown(df):
    total_responses = df.size
    very_satisfied = (df == 'Very Satisfied').sum().sum() / total_responses * 100
    satisfied = (df == 'Satisfied').sum().sum() / total_responses * 100
    not_satisfied = (df == 'Not Satisfied').sum().sum() / total_responses * 100
    very_dissatisfied = (df == 'Very Dissatisfied').sum().sum() / total_responses * 100
    na_or_dont_remember = (df == 'N/A or Don\'t Remember').sum().sum() / total_responses * 100
    
    return pd.Series({
        'Very Satisfied': very_satisfied,
        'Satisfied': satisfied,
        'Not Satisfied': not_satisfied,
        'Very Dissatisfied': very_dissatisfied,
        'N/A or Don\'t Remember': na_or_dont_remember
    })

first_year_breakdown = calculate_satisfaction_breakdown(q9f_data.iloc[:, 1:])
transfer_breakdown = calculate_satisfaction_breakdown(df_cleaned.iloc[:, 1:])

breakdown_df = pd.DataFrame({
    'First-Year': first_year_breakdown,
    'Transfer': transfer_breakdown
})

print(breakdown_df.round(2))

# Plot the breakdown
breakdown_df.plot(kind='bar', figsize=(12, 6))
plt.title('Satisfaction Breakdown: First-Year vs Transfer Students')
plt.xlabel('Satisfaction Level')
plt.ylabel('Percentage')
plt.legend(['First-Year', 'Transfer'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('satisfaction_breakdown.png', dpi=300)
plt.show()
plt.close()

print("\
Satisfaction breakdown plot saved as 'satisfaction_breakdown.png'")

print(satisfaction_df)
```
